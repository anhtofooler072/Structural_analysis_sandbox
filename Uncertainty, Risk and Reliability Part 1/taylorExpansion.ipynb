```python
import numpy as np
import matplotlib.pyplot as plt
import math
import ipywidgets as widgets
from mpl_toolkits.mplot3d import Axes3D     
```

## Approximating with Taylor series:
$$ \ \ln(x) \approx \sum_{n=1}^{\infty} (-1)^{n+1} \frac{(x-1)^n}{n} \$$

```python
def Taylor_ln(x, term):
    approximation = np.zeros_like(x) # Create an array of zeros with the same shape and type as x
    interval = np.linspace(0.5,2,500)
    fy = np.log(interval)
    print(term)
    for i in range(1, term+1):
        term = ((-1)**(i+1) * (x-1)**i) / i
        approximation += term
    plt.clf()     
    plt.figure(figsize=(10, 6))
    plt.plot(interval, fy, label='ln(x)', linewidth=0.75, color='black')  
    plt.plot(interval, approximation, label='Taylor Series Approximation', linestyle='--', alpha=0.7, linewidth=2)
    plt.axvline(x=1, color='red', alpha=0.5, linewidth=0.5)
    plt.text(1.01, 1, 'x=1', color='red', alpha=0.5)
    plt.legend()
    plt.grid(True, alpha=0.2)
    plt.xlim(0.5,2)
    plt.ylim(-2,2)
    plt.show()
    return 0

widgets.interact(Taylor_ln, x=widgets.fixed(np.linspace(0.5,2,500)), term=widgets.IntSlider(min=1, max=50, step=1, value=1))
```

## Taylor series for multivariable functions:

```python
x = np.linspace(-1, 1, 50)
y = np.linspace(-1, 1, 50)

X, Y = np.meshgrid(x, y)

Z_actual = np.exp(X*Y)
Z_approx = 1 + X*Y

fig = plt.figure(figsize=(15, 6))
ax1 = fig.add_subplot(121, projection='3d')
ax2 = fig.add_subplot(122, projection='3d')

plt.colorbar(ax1.plot_surface(X, Y, Z_actual, cmap='viridis', alpha=0.8), ax=ax1, shrink=0.5, aspect=5)
plt.colorbar(ax2.plot_surface(X, Y, Z_approx, cmap='jet', alpha=0.8), ax=ax2, shrink=0.5, aspect=5)
ax1.set_title('Actual Function')
ax2.set_title('Taylor Series Approximation')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.contourf(X, Y, Z_actual-Z_approx, cmap='jet', alpha=0.85)
plt.plot(0, 0, 'ro')
plt.colorbar()
plt.title('Error')

plt.show()
```
